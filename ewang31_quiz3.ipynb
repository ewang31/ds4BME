```
print('Ellen Wang, ewang31@jhu.edu, ewang31')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#Load data
! if [ ! -e orings.csv ] ; \
  then wget https://raw.githubusercontent.com/bcaffo/ds4bme_intro/master/data/orings.csv; \
fi

#Plot o-ring failure vs temperature
df = pd.read_csv("orings.csv")
L = df['Total'].values
L[L>1]=1
temp = df['Temperature'].values   
plt.scatter(temp,L)
plt.xlabel('Temperature')
plt.ylabel('O-ring failure probability')

#Logistic regression fit
from sklearn.linear_model import LogisticRegression
temp = temp.reshape(-1,1)
L = L.reshape(-1,1)
lr = LogisticRegression(fit_intercept = True, solver = 'lbfgs')
lr.fit(temp,L)
print('b0 = {:.3f}\nb1 = {:.3f}'.format(lr.intercept_[0], lr.coef_[0][0]))

#Plot logistic fit
b0,b1 = [lr.intercept_[0],lr.coef_[0][0]]
x = np.linspace(50,85,100)
sigmoid_func = np.exp(b0+b1*x)/(1+np.exp(b0+b1*x))
plt.plot(x,sigmoid_func)

#Estimated probability
p = np.exp(b0+b1*74)/(1+np.exp(b0+b1*74))
print('The probability of a new oring failure at 74 degrees is {:.3f}.'.format(p))


```
